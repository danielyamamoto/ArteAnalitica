```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avocado = pd.read_csv("avocado.csv")
avocado.head()
```

```python
avocado.describe()
```

```python
data = avocado.corr()
data
```

```python
# BOXPLOT WITH MATPLOT
ax = sns.heatmap(data)
plt.show()
```

```python
col1 = avocado['AveragePrice'] 
fig = plt.figure(figsize =(10, 7)) 
plt.boxplot(col1)
plt.show()
```

```python
col1 = avocado['year'] 
fig = plt.figure(figsize =(10, 7)) 
plt.boxplot(col1)
plt.show()
```
